```python
from collections import Counter
```

```python
def get_values(df, attrib):
    count = Counter()
    for v in df[attrib]:
        count[v] += 1
    return count
```

```python
count = get_values(df, col)
count
```

```python
def get_data_skeleton(df):
    dataskeleton = {}
    i = 0
    for col in df.columns:
        dataskeleton[i] = get_values(df, col)
        i += 1
    return dataskeleton
```

```python
dataskeleton = get_data_skeleton(df)
```
